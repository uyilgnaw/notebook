```python
class Test1:
    def f1(self):
        self.name="张三"
        self.__age=20
        print(self.name)
        print(self.__age)
        
class Test2(Test1):
    def f(self):
        print(self.name)
        print(self.__age)
        
inst1=Test1()
inst1.f1()

Test2.f(inst1)

print(Test1.__dict__)
# 此处更改只是临时改变
inst1._Test1__age=21
print(inst1._Test1__age)
inst1.f1()
```

```python
import collections

s = [('yellow',1),('blue',2),('yellow',3),('blue',4),('red',1)]

d = collections.defaultdict(list)

for k,v in s:
    print(k,v)
    d[k].append(v)
print(d)
print(d.keys())
print(list(d.items()))
```

```python
import collections

s = [('yellow',1),('blue',2),('yellow',3),('blue',4),('red',1)]

# defaultdict
d = collections.defaultdict(list)

for k, v in s:
    d[k].append(v)
    
# use dict and setdefault

g = {}

for k ,v in s:
    g.setdefault(k,[]).append(v)
    
e = {}
for k,v in s :
    e[k] = v

print(d.items())

print(g.items())

print(e.items())

```
